```
data=data.lower()
```

```
z=len(data)
k=3
```

```
t={}
for i in range(z-k):

 x=data[i:k+i]
 y=data[k+i]
 if t.get(x) is None:
  t[x]={}
  t[x][y]=1
 else: 
  if t[x].get(y) is None:
   t[x][y]=1
  else: 
   t[x][y]+=1
print(t)
```

```
inp=input()

for i in range(10000):
    iv=inp[-3:]
    pc=np.array(list(t[iv].keys()))
    pv=np.array(list(t[iv].values()))
    s=sum(t[iv].values())
    pv=np.array(pv)/s
    inp=inp+np.random.choice(pc,p=pv)
print(inp)    
    
```
